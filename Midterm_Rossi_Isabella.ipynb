```python
#histogram for element 1
import matplotlib.pyplot as plt

#defining the number list
numbers_list = [
    2,4,6,8,4,5,2,1,9,0,4,6,7,4,3,2,1,9,10,3,7,9,6,0,1,3,5,6,7,8,9,10,2,3,6,8,9,10,6,7,4,3
]

#couonting the frequencies of the numbers with dictionary
frequency_dict = {}
for number in numbers_list:
    if number in frequency_dict:
        frequency_dict[number] += 1
    else:
        frequency_dict[number] = 1

#ploting  frequency results using histogram
plt.hist(numbers_list, bins=range(11))
plt.title('Frequency of Numbers')
plt.xlabel('Number')
plt.ylabel('Frequency')
plt.hist(numbers_list, color = "pink", ec="red")
plt.show()

print(frequency_dict)
```

```python
import pandas as pd

df = pd.read_csv('amazon-orders.csv')

#fixing the $ character with a blank space and convert to float
df['Total Charged'] = df['Total Charged'].str.replace('$', '').astype(float)

print(df.dtypes)
print(df.head())
```

```python
average = df['Total Charged'].mean()
median = df['Total Charged'].median()
maximum = df['Total Charged'].max()
minimum = df['Total Charged'].min()
std_dev = df['Total Charged'].std()

print(f"Average: {average}")
print(f"Median: {median}")
print(f"Maximum: {maximum}")
print(f"Minimum: {minimum}")
print(f"Standard Deviation: {std_dev}")
```

```python
#based on the price ranges of the Total Charged
#defining the intervals for the pie chart
bins = range(0, 201, 20)

groups = pd.cut(df["Total Charged"], bins=bins, include_lowest=True)

counts = groups.value_counts()

plt.pie(counts, labels=counts.index.astype(str), autopct="%1.1f%%")
plt.title("Total Charged by Intervals of 20")
plt.show()
```

```python
#reating a bar chart of Tax Charged
plt.bar(df.index, df['Tax Charged'])

plt.title('Tax Charged')
plt.xlabel('Index')
plt.ylabel('Tax Charged')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('amazon-orders.csv')

#converting columns to float
df['Subtotal'] = df['Subtotal'].str.replace('$', '').astype(float)
df['Total Charged'] = df['Total Charged'].str.replace('$', '').astype(float)

df['Order Date'] = pd.to_datetime(df['Order Date'])
grouped = df.groupby(df['Order Date'].dt.month)

means = grouped.mean()[['Subtotal', 'Total Charged']]

#making a line chart
ax = means.plot(kind='line', figsize=(8,6), color=['hotpink', 'green'])

ax.set_xlabel('Month')
ax.set_ylabel('Amount ($)')
ax.set_title('Subtotal and Total Charged Comparison')

# Show the chart
plt.show()


plt.show()
```
